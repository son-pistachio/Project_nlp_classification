```python
# 필요한 라이브러리 임포트
import pandas as pd
from transformers import BertTokenizerFast
import matplotlib.pyplot as plt

# 데이터 로드
df = pd.read_csv("/home/son/ml/nlp_classification/datasets/final_data.csv")

# 토크나이저 초기화
tokenizer = BertTokenizerFast.from_pretrained('kykim/bert-kor-base')

# 각 문장의 토큰 길이 계산
token_lengths = df['text'].apply(lambda x: len(tokenizer.tokenize(x)))

# 토큰 길이 분포 히스토그램 그리기
plt.figure(figsize=(10,6))
plt.hist(token_lengths, bins=50, color='skyblue', edgecolor='black')
plt.title('Token length distribution')
plt.xlabel('# tokens')
plt.ylabel('# sentences')
plt.show()
```

```python
# 필요한 라이브러리 임포트
import pandas as pd
from transformers import BertTokenizerFast
import matplotlib.pyplot as plt

# 데이터 로드
df = pd.read_csv("/home/son/ml/nlp_classification/datasets/final_data.csv")

# 토크나이저 초기화
tokenizer = BertTokenizerFast.from_pretrained('kykim/bert-kor-base')

# 각 문장의 토큰 길이 계산
token_lengths = df['text'].apply(lambda x: len(tokenizer.tokenize(x)))

# 토큰 길이 통계 출력
print("토큰 길이 통계:")
print(f"최소 길이: {token_lengths.min()}")
print(f"최대 길이: {token_lengths.max()}")
print(f"평균 길이: {token_lengths.mean():.2f}")
print(f"중앙값: {token_lengths.median()}")
print(f"표준편차: {token_lengths.std():.2f}")

# 백분위수 출력
print("\n백분위수:")
percentiles = [25, 50, 75, 90, 95, 99]
for p in percentiles:
    value = token_lengths.quantile(p / 100)
    print(f"{p}th 백분위수: {value}")

# 토큰 길이 분포 히스토그램 그리기
plt.figure(figsize=(10, 6))
plt.hist(token_lengths, bins=50, color='skyblue', edgecolor='black')
plt.title('문장 토큰 길이 분포')
plt.xlabel('토큰 개수')
plt.ylabel('문장 수')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
